```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import bernoulli

def aBernoulliExperiment(randomNumGen, bernoulliParam_p=0.5, numRepeatsOfExp=1, numTrialsInAnExp=1, profitLoss=[1, 0]):
    observation = np.ones((numRepeatsOfExp, numTrialsInAnExp))
    observation = randomNumGen.random((numRepeatsOfExp, numTrialsInAnExp))
    index_ge1 = np.less_equal(observation, bernoulliParam_p)
    observation[index_ge1] = profitLoss[0]
    observation[np.invert(index_ge1)] = profitLoss[1]
    return observation

# We toss the coin 20 times and plot the outcomes.
v_numTrialsInAnExp = 20
observation = aBernoulliExperiment(np.random.default_rng(), bernoulliParam_p=0.2, numRepeatsOfExp=1, numTrialsInAnExp=v_numTrialsInAnExp, profitLoss=[1,0])
plt.figure()
plt.step(range(1, v_numTrialsInAnExp+1), observation[0], where='mid')
plt.xlabel("Trial Number")
plt.ylabel("Outcome (1 for Head, 0 for Tail)")
plt.title("Outcome of Single Coin Tosses")
plt.tight_layout()



```

```python
import matplotlib.pyplot as plt
import numpy as np

def aBernoulliExperiment(randomNumGen, bernoulliParam_p=0.2, numRepeatsOfExp=1, numTrialsInAnExp=1, profitLoss=[1, 0]):
    observation = randomNumGen.random((numRepeatsOfExp, numTrialsInAnExp))
    index_ge1 = np.less_equal(observation, bernoulliParam_p)
    observation[index_ge1] = profitLoss[0]
    observation[np.invert(index_ge1)] = profitLoss[1]
    return observation

# Repeat (a) 100 times and plot the histogram of the total number of 1s in each sequence.
numRepeats = 100
v_numTrialsInAnExp = 20  # Assuming this is the number of trials per experiment
observations = aBernoulliExperiment(np.random.default_rng(), bernoulliParam_p=0.2, numRepeatsOfExp=numRepeats, numTrialsInAnExp=v_numTrialsInAnExp, profitLoss=[1,0])
total_ones = np.sum(observations, axis=1)

# Choose all those sequences that sum to at least 3.
selected_sequences = [seq for seq in total_ones if seq >= 3]

plt.figure()
plt.hist(selected_sequences, bins=np.arange(v_numTrialsInAnExp+1), density=True, align='left', rwidth=0.8, color='orange', alpha=0.8)
plt.xlabel("Total Number of 1s")
plt.ylabel("Frequency")
plt.title("Histogram of Total Number of 1s in Sequences Summing to at Least 3")
plt.show()
```

This is very similar part (c) but only difference is we want sum of 1's to be atleast 3, thus values of total number of 1's less than 3 is not plotted. In this case also the expected value can be found by n*p which is roughly 0.4, that is, getting four 1's has highest chance. Also, due to less probability of p total number of 1's is generally less. This graph also shows us the randomness of the total number of 1's but mostly it frequents in the range of total of 3-5 one's.

```python
import numpy as np
import matplotlib.pyplot as plt

num_of_experiments = 10000

alphas = np.random.uniform(0, 100, size=4)

for alpha in alphas:
    poisson_values = np.random.poisson(alpha, num_of_experiments)

    # here we predefined n and p for simplicity, but randomized alpha
    n = 2 * alpha
    p = 0.5
    binomial_values = np.random.binomial(n, p, num_of_experiments)

    plt.figure(figsize=(8, 6))
    plt.hist(poisson_values, bins=20, density=True, alpha=0.5, color='blue', label=f'Poisson({alpha})')
    plt.hist(binomial_values, bins=20, density=True, alpha=0.5, color='red', label=f'Binomial({n}, {p})')
    
    plt.title(f'Poisson vs. Binomial for alpha = {alpha}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

In this part, we first randomly took alpha and for simplicity, we pre defined values for n and p. Then we used this data to plot a binomial and Poisson distribution to compare the two different types of distriibution. As we know, the expected value of binomial is is generally more than poisson and that is what is mostly shown in the above graphs. The different and random values of alpha showed the changing distribution of graphs and one can also observe wierd shaped graphs if alpha is small.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Choose two values of alpha randomly
values_of_alpha = np.random.uniform(1, 20, size=2)

N = [5, 10, 100, 1000]

for alpha in values_of_alpha:
    plt.figure(figsize=(20, 8))
    for i, num_of_RVs in enumerate(N):
        plt.subplot(1, len(N), i + 1)

        # Poisson RVs
        poisson_sequence = np.sum(np.random.poisson(lam=alpha, size=(1000, num_of_RVs)), axis=1)
        plt.hist(poisson_sequence, bins=30, density=True, alpha=0.5, color='green')
        plt.title(f'alpha = {alpha:.2f}, N = {num_of_RVs}')

    plt.tight_layout()
    plt.show()
```

This code randomly picks 2 different alphas and for all the 4 given values of N, it plots graphs with each alpha. Importanty it demonstrates how as the number of trials increase, we get a bell shaped curve (gaussian type) which is the main aspect of the central limit theorem. This experiment plotted 8 different graphs using varing parameters of N and alpha. Upon using such a large number of experiment, we can not onlt prove the central limit theorem, but also the law of natural numbers.
